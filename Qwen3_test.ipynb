```python
from huggingface_hub import hf_hub_download,snapshot_download
import os

def hf_download(model_name_or_path,model_basename,local_path,tokens):
    if not os.path.exists(local_path):
        os.mkdir(local_path)
    #Download the model from repo
    if len(model_basename.strip())>0:
        if len(tokens.strip())>0:
            model_path = hf_hub_download(repo_id = model_name_or_path,filename = model_basename, cache_dir = local_path,token=tokens)
            print("Download Completed.")
        else:
            model_path = hf_hub_download(repo_id = model_name_or_path,filename = model_basename, cache_dir = local_path)
            print("Download Completed.")
    else:
        if len(tokens)>0:
            model_path = snapshot_download(repo_id = model_name_or_path, cache_dir = local_path,token=tokens)
            print("Download Completed.")
        else:
            model_path = snapshot_download(repo_id = model_name_or_path, cache_dir = local_path)
            print("Download Completed.")
    return model_path
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer,set_seed,pipeline

model_name = "Qwen/Qwen3-4B"

# load the tokenizer and the model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto",
    device_map="auto"
)

# prepare the model input
prompt = "Give me a short introduction to large language model."
messages = [
    {"role": "user", "content": prompt}
]
text = tokenizer.apply_chat_template(
    messages,
    tokenize=False,
    add_generation_prompt=True,
    enable_thinking=True # Switches between thinking and non-thinking modes. Default is True.
)
```

```python
import torch
```

```python
generator = pipeline('text-generation',model=model,tokenizer=tokenizer,
                     return_full_text=False,max_new_tokens=750,temperature=0.6, top_p=0.95, top_k=20, min_p=0)
with torch.no_grad():
  set_seed(42)
  res = generator(text)[0]['generated_text']
```

```python
print(res)
```

```python
import shutil
import os

def force_delete_folder(folder_path):
    try:
        shutil.rmtree(folder_path)
        print(f"Folder '{folder_path}' and its contents deleted successfully.")
    except FileNotFoundError:
        print(f"Folder '{folder_path}' not found.")
    except Exception as e:
        print(f"Error deleting folder '{folder_path}': {e}")

# Example usage:
folder_to_delete = "local"
force_delete_folder(folder_to_delete)
```
